```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_excel('tmp.xlsx')
df = df.iloc[2:-1]
```

```python
#renaming all columns
df = df.rename(columns={
    'Sygnatura czasowa': 'timestamp',
    'Age': 'age',
    'Gender': 'gender',
    'Sexual orientation': 'sex_orient',
    'Relationship status': 'rel_status',
    'Did you have a father figure growing up?': 'fath_fig',
    'Type of person': 'person_type',
    'On a scale of 1 - 7, how agreeable do you think you are?': 'agreeable_scale',
    'On a scale of 1 - 7, how self-aware do you think you are?': 'self_aware_scale',
    'On a scale of 1 - 7, how emotionally stable do you think you are?': 'emotionally_stable_scale',
    'On a scale of 1 - 7, how emotionally stable do you think you are?': 'emotionally_stable_scale',
    'On a scale of 1 - 7, how much are you open to new experiences?': 'new_exp_scale',
    'Do you find it hard to communicate with other people?': 'hard_comm_ppl',
    'Do you find it hard to understand how other people think or feel?': 'understand_ppl_feel',
    'Do you get anxious in social situations?': 'anxious_situations',
    'What is your highest education': 'education',
    'What is your employment status?': 'work_status',
    'If you\'re working, please tell us what type of job are you doing?': 'work_type',
    'On a scale of 1 - 7, how much do you enjoy hanging out with friends outside?  ': 'friends_hangout_scale',
    'What social media platform are you using the most?': 'social_most',
    'With what age group of people do you prefer to interact?': 'ppl_group_prefer',
    'How much do you use your smartphone per day?': 'phone_usage',
    'How much do you use your computer?': 'pc_usage',
    'What type of media are you consuming?': 'socials_consuming',
    'What are you most passionate about? ': 'passion',
    'What type of video games do you enjoy most? ': 'fav_games',
    'Do you like playing board games?': 'like_boardgames',
    'What type of movies are generally your favorite?': 'fav_movies',
    'What type of books are you reading the most?': 'fav_books',
    'What subculture you feel you belong to?': 'subculture'
})
```

```python
mapper = {
 'netflix': 'Netflix',
 'other_str': 'Other streaming services',
 'streams': 'Stream content',
 'tv': 'Television',
 'tiktok': 'TikTok',
 'yt': 'Youtube'}

for key, value in mapper.items():
    df[f'socials_consuming_{key}'] = df[['socials_consuming']].apply(lambda x: int(value in x['socials_consuming']), axis=1)
```

```python
def class_mapper(x):
    return int(x == 'Furry')

df['subculture'] = df[['subculture']].apply(lambda x: class_mapper(x), axis=1)
```

```python
grp = df[['timestamp','fav_books']].groupby(['fav_books'])
a = grp['timestamp'].count().reset_index()
most_books = a[a['timestamp'] >= 3]\
    .sort_values('timestamp', ascending=False)\
    ['fav_books'].head(5).unique()

def parse_books(x):
    if x in most_books:
        return x
    if 'none' in x or 'don\'t' in x:
        return 'none'
    return 'Other'

uniq_befor = df['fav_books'].unique()

df['fav_books'] = df['fav_books'].fillna('none')
df['fav_books'] = df.apply(lambda x: parse_books(x['fav_books']), axis=1)

print(uniq_befor)
print(df['fav_books'].unique())
```

```python
df['fav_movies'] = df['fav_movies'].fillna('none')
grp = df[['timestamp','fav_movies']].groupby(['fav_movies'])
a = grp['timestamp'].count().reset_index()
most = a[a['timestamp'] > 2]\
    .sort_values('timestamp', ascending=False)\
    ['fav_movies'].unique()

print(df['fav_movies'].unique())
print(most)

def parse_movies(x):
    if x in most:
        return x
    if 'none' in x:
        return 'none'
    return 'Other'

df['fav_movies'] = df.apply(lambda x: parse_movies(x['fav_movies']), axis=1)
print(df['fav_movies'].unique())
```

```python
df['fav_games'] = df['fav_games'].fillna('none')
grp = df[['timestamp','fav_games']].groupby(['fav_games'])
a = grp['timestamp'].count().reset_index()
most = a[a['timestamp'] > 4]\
    .sort_values('timestamp', ascending=False)\
    ['fav_games'].unique()

print(df['fav_games'].unique())
print(most)

def parse_games(x):
    if x in most:
        return x
    if 'none' in x:
        return 'none'
    return 'Other'

df['fav_games'] = df.apply(lambda x: parse_games(x['fav_games']), axis=1)
print(df['fav_games'].unique())
```

```python
print(df['like_boardgames'].unique())

def parse_board(x):
    return int(x == 'Yes' or x == 'I love board games')

df['like_boardgames'] = df.apply(lambda x: parse_board(x['like_boardgames']), axis=1)
df['like_boardgames'].unique()
```

```python
df['passion'] = df['passion'].fillna('none')
grp = df[['timestamp','passion']].groupby(['passion'])
a = grp['timestamp'].count().reset_index()
most = a[a['timestamp'] > 4]\
    .sort_values('timestamp', ascending=False)\
    ['passion'].unique()

print(df['passion'].unique())
print(most)

def parse_passion(x):
    if x in most:
        return x
    if x == 'Music, Photography and art, Technology and Gaming, Traveling and Science':
        return 'I am not really passionate about one thing'
    if x == 'Slavic mithology':
        return 'Reading and Literature (e.g., novels, poetry, non-fiction)'
    return 'Other'

df['passion'] = df.apply(lambda x: parse_passion(x['passion']), axis=1)
df['passion'].unique()
```

```python
def parse_pc_usage(x):
    if x in ['1 - 2h', '< 1h']:
        return '< 2h'
    return x

print(df['pc_usage'].unique())

df['pc_usage'] = df.apply(lambda x: parse_pc_usage(x['pc_usage']), axis=1)
print(df['pc_usage'].unique())
```

```python
print(df['ppl_group_prefer'].unique())

def parse(x):
    if x in ['25 - 30', '30 - 45', '> 45']:
        return '> 25'
    return x

df['ppl_group_prefer'] = df.apply(lambda x: parse(x['ppl_group_prefer']), axis=1)
print(df['ppl_group_prefer'].unique())
```

```python
print(df['phone_usage'].unique())

def parse(x):
    if x in ['1 - 2h', '< 1h']:
        return '> 2h'
    if x in ['more than 8h', '6 - 8h']:
        return 'more than 6h'
    return x

df['phone_usage'] = df.apply(lambda x: parse(x['phone_usage']), axis=1)
print(df['phone_usage'].unique())
```

```python
print(df['work_type'].unique())

grp = df[['timestamp','work_type']].groupby(['work_type'])
a = grp['timestamp'].count().reset_index()
most = a[a['timestamp'] > 2]\
    .sort_values('timestamp', ascending=False)\
    ['work_type'].unique()

def parse(x):
    if x in most:
        return x
    if 'student' in x or \
        'Studying' in x or \
        '-' in x.lower() or \
        'not' in x.lower() or \
        'unemployed' in x.lower() or \
       x.lower() in [
        'no', 'none', 'uwu', 'A',
        'Crippled for life'
    ]:
        return 'not working'
    if x in [
        'Operator linii produkcyjnej części samochodowych',
        'locksmith/electrician',
        'induction loop technician',
        'manufacturing'
    ]:
        return 'technical work'
    if x in [
        'Paramedic / ambulance driver',
        'Military', 'Hospitality '
    ]:
        return 'public services'
    if x == 'scientist':
        return 'education-related job'
    return 'Other'

df['work_type'] = df.apply(lambda x: parse(x['work_type']), axis=1)
print(df['work_type'].unique())
```

```python
print(df['fath_fig'].unique())

df['fath_fig'] = df.apply(lambda x: int('Yes' in x['fath_fig']), axis=1)
print(df['fath_fig'].unique())
```

```python
print(df['age'].unique())

def parse(x):
    if x in ['< 14 yo', '15 - 17 yo']:
        return '< 18 yo'
    if x in ['27 - 29 yo', '24 - 26 yo', '30 - 32 yo', '> 36 yo']:
        return '> 23 yo'
    return x

df['age'] = df.apply(lambda x: parse(x['age']), axis=1)
print(df['age'].unique())
```

```python
print(df['rel_status'].unique())

def parse(x):
    if x in['engaged', 'married']:
        return 'commited'
    return x

df['rel_status'] = df.apply(lambda x: parse(x['rel_status']), axis=1)
print(df['rel_status'].unique())
```

```python
excluded = ['socials_consuming', 'like_boardgames', 'fath_fig', 'timestamp', 'subculture']
excluded = ['socials_consuming', 'timestamp']
features = [col for col in df.columns if col not in excluded and 'socials_consuming' not in col]
df1 = pd.get_dummies(df, columns=features)
df1 = df1[[col for col in df1.columns if col not in ['socials_consuming', 'timestamp'] and 'socials_consuming' not in col]]
print(df1.dtypes)
df1.head(1)
```

```python
nots = [col for col in df1.columns if 'none' in col or 'not working' in col]
df1 = df1[[col for col in df1.columns if col not in nots]]
```
